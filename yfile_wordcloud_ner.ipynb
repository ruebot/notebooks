```python
example_file = open('corpus/2021-10-03-national-day-for-truth-and-reconciliation-panel-focuses-on-reconciliation-in-action.txt')
words = example_file.read()
words
```

```python
wordcloud = WordCloud(
    width=2000,
    height=1500,
    scale=10,
    max_font_size=250,
    max_words=100,
    background_color="white",
).generate(words)
plt.figure(figsize=[35, 10])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title(words.partition('\n')[0], fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.show()
```
