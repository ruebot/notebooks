# Tweets to @realDonaldTrump hashtags (20201225 - 20210114)

We're using a Jupyter Notebook via the [Anaconda distribution](https://www.anaconda.com/), and we'll need to only import the following:

````python
import pandas as pd
````

We'll create a DataFrame to work with from `202101-hashtags.txt`. This file is a derivative of the [Tweets to Donald Trump (@realDonaldTrump)](https://doi.org/10.5683/SP/8BAVQM) dataset. It can be created with the [`tags.py`](https://github.com/DocNow/twarc/blob/master/utils/tags.py):

```
$ python twarc/utils/tags.py 202101-tweets.jsonl > 202101-hashtags.txt
```

It is also included in the repo in the `data` directory.

````python
hashtags = pd.read_csv(
    "data/202101-hashtags.txt",
    sep=" ",
    names=["count", "hashtag"],
    header=None,
    encoding='utf8',
)
````

Let's take a look at the DataFrame:

````python
hashtags
````

The data source is already sorted, so we can move right to plotting it.

We're looking at 73,516 unique hashtags in the dataset, so we don't want to plot all of them. Let's work with the top 20. We can take advantage of the `head()` function here to limit our DataFrame results.

Let's also do a colour gradient for the bars to get some color separation.

Here we'll create the chart using the `head()` function, then add a title, and labels.

````python
tags = hashtags.head(10)
  
colours = [(x / 100.0, x / 20.0, 0.30) for x in range(len(tags))]

chart = tags.plot.bar(
    color=colours,
    x="hashtag",
    y="count",
    figsize=(35, 20),
    legend=False,
    edgecolor="pink",
)

chart.set_title("Tweets to @realDonaldTrump hashtags (20201225 - 20210114)", fontsize=50)
chart.set_xlabel("Hashtag", fontsize=30)
chart.set_ylabel("Count", fontsize=30)
````
