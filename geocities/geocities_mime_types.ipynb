```python
data_tika = "https://ruebot.net/mime-tika.csv"
data_ws = "https://ruebot.net/mime-ws.csv"
```

```python
import pandas as pd
```

```python
tika = pd.read_csv(data_tika)
tika
```

```python
ws = pd.read_csv(data_ws)
ws
```

```python
import altair as alt
```

```python
tika_base = (
    alt.Chart(tika.head(10))
    .mark_arc(innerRadius=200)
    .encode(
        theta=alt.Theta(field="Count", type="quantitative", stack=True),
        color=alt.Color(field="MIME Type", type="nominal"),
    )
  )

tika_base.properties(
    width=1100, height=1000, title="Geocities File Format Distribution (Tika)"
)
```

```python
ws_base = (
    alt.Chart(ws.head(10))
    .mark_arc(innerRadius=200)
    .encode(
        theta=alt.Theta(field="Count", type="quantitative", stack=True),
        color=alt.Color(field="MIME Type", type="nominal"),
    )
  )

ws_base.properties(
    width=1100, height=1000, title="Geocities File Format Distribution (Web Server)"
)
```
