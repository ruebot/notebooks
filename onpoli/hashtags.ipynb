# #onpoli tweet volume July 14, 2009 - December 31, 2022

```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv(
    "data/onpoli-hashtags.csv",
    names=["Hashtag", "Count"],
    header=None
)
```

```python
df = df.iloc[1:]
```

```python
df.head(25)
```

```python
chart = alt.Chart(df.head(10)).mark_bar().encode(
               x=alt.X("Hashtag:O", sort="-y"),
               y=alt.Y("Count:Q")
           )

chart_text = chart.mark_text(align="center", baseline="bottom").encode(
     text="Count:Q"
)

(chart + chart_text).properties(
    width=1500,
    height=2200,
    title="#onpoli Hashtags July 14, 2009 - December 31, 2022",
).configure_title(fontSize=30, font="Verdana", align="center", color="black", dy=-3)
```
