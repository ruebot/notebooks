# Graphing #elxn43 hashtags

We're using a Jupyter Notebook via the [Anaconda distribution](https://www.anaconda.com/), and we'll need to only import the following:

````python
import pandas as pd
````

We'll create a DataFrame to work with from `elxn43-hashtags.txt`. This file is a derivative of the [#elxn43 dataset](https://doi.org/10.5683/SP2/QAMPPI). It can be created with the [`tags.py`](https://github.com/DocNow/twarc/blob/master/utils/tags.py):

```
$ python twarc/utils/tags.py elxn43_search.jsonl > elxn43-hashtags.txt
```

It is also included in the repo in the `data` directory.

````python
elxn43_hashtags = pd.read_csv(
    "data/elxn43-hashtags.txt",
    sep=" ",
    names=["count", "hashtag"],
    header=None,
    encoding='utf8',
)
````

Let's take a look at the DataFrame:

````python
elxn43_hashtags
````

The data source is already sorted, so we can move right to plotting it.

We're looking at 43,626 unique hashtags in the dataset, so we don't want to plot all of them. Let's work with the top 20. We can take advantage of the `head()` function here to limit our DataFrame results.

Let's also do a colour gradient for the bars to get some color separation.

Here we'll create the chart using the `head()` function, then add a title, and labels.

````python
hashtags = elxn43_hashtags.head(20)
  
colours = [(x / 100.0, x / 20.0, 0.30) for x in range(len(hashtags))]

chart = hashtags.plot.bar(
    color=colours,
    x="hashtag",
    y="count",
    figsize=(35, 20),
    legend=False,
    edgecolor="pink",
)

chart.set_title("#elxn43 hashtags", fontsize=50)
chart.set_xlabel("Hashtag", fontsize=20)
chart.set_ylabel("Count", fontsize=20)
````
