# National Day for Truth and Reconciliation panel explores reconciliation in action

```python
data = open('corpus/2021-10-03-national-day-for-truth-and-reconciliation-panel-focuses-on-reconciliation-in-action.txt')
article = data.read()
article
```

```python
wordcloud = WordCloud(
    width=2000,
    height=1500,
    scale=10,
    max_font_size=250,
    max_words=100,
    background_color="white",
).generate(article)
plt.figure(figsize=[35, 10])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title(article.partition('\n')[0], fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.show()
```

```python
%%capture
!pip install spacy
```

```python
import en_core_web_sm
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

nlp.max_length = 1100000
```

```python
ner = nlp(article)
displacy.render(ner, style="ent", jupyter=True)
```
